1. First clean the data before process

```python
import os
import codecs
import re

directory = '../Task/texts/'

def clean_text(text):
    # Remove unwanted characters
    text = re.sub(r"[,.;?!\"-]", "", text)
    # Optional: Convert to lowercase
    text = text.lower()
    # Optional: Remove extra white spaces
    text = re.sub(r"\s+", " ", text)
    return text

# Get the list of files in the directory
txt_files = [file for file in os.listdir(directory)]

for file in txt_files:
    file_path = os.path.join(directory, file)
    # Intenta abrir el archivo con diferentes codificaciones hasta que funcione
    for encoding in ['utf-8', 'latin-1', 'utf-16', 'iso-8859-1']:
        try:
            with codecs.open(file_path, 'r', encoding=encoding) as f:
                content = f.read()
                content = content.replace('"','').replace(':','').replace('“','').replace('”','')
                cleaned_content = clean_text(content)
            with open(file_path, 'w', encoding=encoding) as f:
                f.write(cleaned_content)
            break  # Sal del bucle si la lectura y escritura fueron exitosas
        except UnicodeDecodeError:
            continue  # Intenta la siguiente codificación si hay un error de decodificación
```

```python
import os
import pandas as pd
from collections import defaultdict

# Define the path to the directory containing the text files
CORPUS_DIR = '../Task/texts/'
documents = {}

# Dictionary to store word counts for each document
word_counts = defaultdict(lambda: defaultdict(int))
```

```python
# Iterate through each file in the directory
for filename in os.listdir(CORPUS_DIR):
    file_path = os.path.join(CORPUS_DIR, filename)
    with open(file_path, 'r', encoding='latin-1') as file:
        document_text = file.read().lower()  # Read and convert to lowercase
        documents[filename] = document_text

        # Count word occurrences in the current document
        words = document_text.split()
        for word in words:
            word_counts[word][filename] += 1

# Create a DataFrame from the word counts
df = pd.DataFrame(word_counts)

# Fill NaN values with 0 (for words not present in some documents)
df.fillna(0, inplace=True)
```

```python
import os
import pandas as pd
from collections import defaultdict

# Define the path to the directory containing the text files
QUERIS_DIR = '../Task/Queris/'
documents = {}

# Dictionary to store word counts for each document
word_counts = defaultdict(lambda: defaultdict(int))
```

```python
# Iterate through each file in the directory
for filename in os.listdir(QUERIS_DIR):
    file_path = os.path.join(QUERIS_DIR, filename)
    with open(file_path, 'r', encoding='utf-8') as file:
        document_text = file.read().lower()  # Read and convert to lowercase
        documents[filename] = document_text

        # Count word occurrences in the current document
        words = document_text.split()
        for word in words:
            word_counts[word][filename] += 1

# Create a DataFrame from the word counts
df_queris = pd.DataFrame(word_counts)

# Fill NaN values with 0 (for words not present in some documents)
df_queris.fillna(0, inplace=True)
```

```python
import numpy as np
# Lista para almacenar las similitudes parciales
partial_similarities = []

for i in df:
    for j in  df_queris:
        if i == j:
            dot_product = np.dot(df[i], df_queris[j])
            # Calcular las normas de A y B
            norm_df = np.linalg.norm(df[i])
            norm_df_queris = np.linalg.norm(df_queris[j])

            # Calcular la similitud coseno
            cosine_similarity = dot_product / (norm_df * norm_df_queris)
            # Almacenar la similitud parcial en la lista
            partial_similarities.append(cosine_similarity)

    
        similarity_df = pd.DataFrame([partial_similarities] ) # Lista de listas para cada fila del DataFrame
```
